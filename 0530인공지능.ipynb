```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = plt.imread("PCA_1.jpg")
img2 = plt.imread("PCA_2.jpg")
img3 = plt.imread("PCA_3.jpg")
rows,cols,colors = img1.shape
plt.imshow(img3)
```

```python
img_size = rows * cols * colors
```

```python
img1_1D = img1.reshape(img_size)
img2_1D = img2.reshape(img_size)
img3_1D = img3.reshape(img_size)

mean_vec = img1_1D + img2_1D + img3_1D
mean_vec = mean_vec / 3

img1_1D = img1_1D - mean_vec
img2_1D = img2_1D - mean_vec
img3_1D = img3_1D - mean_vec
```

```python
X=np.array([img1_1D , img2_1D, img3_1D])
X = X.T
U,S,Vt = np.linalg.svd(X,full_matrices = False)
print(S)
img1_pca = (1/3) * S[0] **2*U[:,0].reshape(rows,cols,colors)
img2_pca = (1/3) * S[1] **2*U[:,1].reshape(rows,cols,colors)
img3_pca = (1/3) * S[2] **2*U[:,2].reshape(rows,cols,colors)
```

```python
plt.imshow(img3_pca)
print()
```
